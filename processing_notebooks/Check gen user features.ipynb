```python
def get_timestamp_features(vitrine: pd.DataFrame):
    vitrine["lag_timestamp"] = vitrine.groupby(['user_id'])['timestamp'].shift()
    vitrine["chng_timestamp_pred"] = (vitrine["timestamp"] - vitrine["lag_timestamp"]) / vitrine["timestamp"]
    vitrine.drop(["lag_timestamp"], axis = 1, inplace = True)
    return vitrine
```
