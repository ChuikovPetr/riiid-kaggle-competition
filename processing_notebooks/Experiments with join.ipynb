```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', '{:.4f}'.format)
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
%%time
train_mini = pd.read_csv('../data/train_mini.csv')
```

```python
train_mini.head()
```

```python
lectures = pd.read_csv("../data/lectures.csv")
```

```python
lectures = lectures.rename(columns = {'lecture_id': 'content_id', 'tag': 'tags'})
```

```python
lectures
```

```python
lectures['type_of'] = lectures['type_of'].replace('solving question', 'solving_question')
lectures = pd.get_dummies(lectures, columns = ['type_of'], prefix = '', prefix_sep = '')
```

```python
lectures
```

```python
questions = pd.read_csv("../data/questions.csv")
questions = questions.rename(columns = {'question_id': 'content_id'})
questions
```

```python
train_mini.shape
```

```python
train_mini
```

```python
train_mini_lect = train_mini[train_mini['content_type_id'] == 1].reset_index(drop = True)
train_mini_lect = pd.merge(
    train_mini_lect,
    lectures,
    left_on = ['content_id'],
    right_on = ['content_id'],
    how = 'left'
)



train_mini_quest = train_mini[train_mini['content_type_id'] == 0].reset_index(drop = True)
train_mini_quest = pd.merge(
    train_mini_quest,
    questions,
    left_on = ['content_id'],
    right_on = ['content_id'],
    how = 'left'
)
```

```python
train_mini_lect.shape, train_mini_quest.shape
```

```python
train_mini_lect.head()
```

```python
train_mini_quest.head()
```

```python
train_mini_2 = pd.concat([train_mini_lect, train_mini_quest], axis = 0)
train_mini_2 = train_mini_2.sort_values(by = ['user_id', 'timestamp'])
```
